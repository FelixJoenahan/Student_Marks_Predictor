```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## ðŸ“Œ Problem Statement

This project predicts student exam marks based on the number of hours studied using a supervised machine learning algorithm (Linear Regression).

Input: Study Hours  
Output: Predicted Marks  

The goal is to build a regression model that learns the relationship between study time and exam performance.

```python
data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Marks": [30,35,50,55,65,75,85,95]
}
df = pd.DataFrame(data)
df
```

```python
X = df[["Hours"]]   # 2D
y = df["Marks"]     # 1D
```

```python
model = LinearRegression()
model.fit(X, y)

print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

```python
hours = 5.5
pred = model.predict([[hours]])
print(f"Study Hours: {hours}, Predicted Marks: {pred[0]:.2f}")
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks (Linear Regression)")
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

y_pred = model.predict(X)

print("R2 Score:", r2_score(y, y_pred))
print("MAE:", mean_absolute_error(y, y_pred))
```
